```python
import re
text = "Hello 2024 World!"

pattern = r"."
matches = re.findall(pattern, text)
print(matches)
```

```python
text = "Elon phone: 999-111-6666 Tesla CFO: (999)-333-7777"
pattern = r"\d{3}-\d{3}-\d{4}"
print(re.findall(pattern, text))
```

```python
text = "asjl15; kj; alsj dF"
pattern = r"[A-Za-z]+"
print(re.findall(pattern, text))
```

```python
text = """Note 1—Overview
Note 2—Summary of Significant Accounting Policies"""
pattern = r"Note\s+\d+\s*[–—-]\s*(.+)"
print(re.findall(pattern, text))
```

```python
text = """FY2021 Q1 cost was high.
Prior quarter FY2020 Q4 was lower."""
pattern = r"FY\d{4} Q[1-4]"
print(re.findall(pattern, text))
```

```python
text = """test@gmail.com hi xyz@yahoo.com"""
pattern = r"[A-Za-z0-9._%+-]+@(gmail\.com|yahoo\.com)"
print(re.findall(pattern, text))
```

```python
text = """test@gmail.com hi xyz@yahoo.com"""
pattern = r"[A-Za-z0-9._%+-]+@(?:gmail\.com|yahoo\.com)"
print(re.findall(pattern, text))
```

```python
text = "2024-12-31"
pattern = r"\d{4}-\d{2}-\d{2}"
print(re.findall(pattern, text))
```

```python
text = "$4.85"
pattern = r"\$\d+(?:\.\d+)?"
print(re.findall(pattern, text))
```

```python
text = "Visit https://google.com"
pattern = r"https?://[A-Za-z0-9]+\.[A-Za-z]+"
print(re.findall(pattern, text))
```

```python
text = "#python #tesla"
pattern = r"\#[A-Za-z0-9]+"
print(re.findall(pattern, text))
```
