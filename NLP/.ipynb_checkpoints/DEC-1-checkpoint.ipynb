```python
# list comprehension
numbers = [i*i for i in range(1, 6)]
print(numbers)
```

```python
for i in range(1,6):
    print(i*i,end=" ")
```

```python
# list comprehension to print even num
number=[i for i in range(0,51) if(i%2==0)]
print (number)
```
