```python
import re
text = '''
Follow our leader Elon musk on twitter here: https://twitter.com/elonmusk, more information 
on Tesla's products can be found at https://www.tesla.com/. Also here are leading influencers 
for tesla related news,
https://twitter.com/teslarati
https://twitter.com/dummy_tesla
https://twitter.com/dummy_2_tesla
'''
pattern = r'https?://[A-Za-z]+\.[A-Za-z]+(.*)'

re.findall(pattern, text)
```

```python
import re
text = "cat is sittind with another cat"
print(re.findall(r"cat", text))
```

```python
text = "cat cot cut c@t"
print(re.findall(r"c.t",text)) #. matches any one character (except newline)
```

```python
text = "hello"
print(re.findall(r"[aeiou]", text)) #[aeiou] Match any vowel
```

```python
text = "AbC123"
print(re.findall(r"[A-Z]", text))
```

```python
text = "abc123"
print(re.findall(r"[^0-9]", text))
```

```python
text = "User_1 age 24"
print(re.findall(r"\w+", text))
```

```python
text = "Call 9876543210"
print(re.findall(r"\d{10}", text))
```

```python
text = "http://a.com https://b.com"
print(re.findall(r"https?", text))
```

```python
text = "2024-12-31"
print(re.findall(r"(\d{4})-(\d{2})-(\d{2})", text))
```

```python
text = "$4.85 $10"
print(re.findall(r"\$\d+(?:\.\d+)?", text))
```

```python
text = "cat dog cow"
print(re.findall(r"cat|dog", text))
```

```python
text = "Born Elon\nAge 54"
print(re.findall(r"^Born(.*)", text, re.MULTILINE)) #^Born Matches only at start of line
```

```python
text = "cat scatter"
print(re.findall(r"\bcat\b", text))
```

```python
text = "$4.85"
print(re.findall(r"\$\d+\.\d+", text))
```

```python
text = "<h1>Title</h1>"
print(re.findall(r"<.*?>", text))
```
