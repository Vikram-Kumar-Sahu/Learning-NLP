```python
def get_pattern_matches(pattern, text):
        matches = re.findall(pattern, text)
        if matches:
            return matches[0]
```

```python
def get_personal_information(chat):
    age = get_pattern_matches(r'age (\d+)', chat)
    full_name = get_pattern_matches(r'Born(.*)\n', chat)

    return {
        'age': int(age) if age else None,
        'name': full_name.strip() if full_name else None
    }
```

```python
get_personal_information(chat)
```
