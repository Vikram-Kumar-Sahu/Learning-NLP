```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")

doc = nlp("Dr. Strange loves pav bhaji of mumbai. Hulk loves chaat of delhi")

for sentence in doc.sents:
    print(sentence)
```

```python
for sentence in doc.sents:
    for word in sentence:
        print(word)
```
